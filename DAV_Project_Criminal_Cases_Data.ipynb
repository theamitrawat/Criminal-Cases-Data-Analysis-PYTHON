# Importing Required Packages

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.offline as py
import plotly.graph_objs as go
import plotly.tools as tls
import plotly.express as px
```

### Reading dataset
### (source url = https://ndap.niti.gov.in/dataset/7150)

```python
court = pd.read_csv('NDAP_REPORT_7150.csv')
court.head(10)
```

### Removing unnecessary rows and columns

```python
df = court.drop(['Rowid','Country','State lgd code','District lgd code','Year'],axis=1)
df.head(10)
```

```python
court_df = df[df['District and taluk court case type']!='Total']

court_df.head(10)
```

### Renaming the columns

```python
court_df = court_df.rename(columns={'District and taluk court case type':'Court_case_type'})

court_df.head(10)
```

### Grouping the dataset on the basis of case type (criminal and civil cases)

```python

court.groupby('District and taluk court case type')['Pending cases'].agg('sum')/10000000
```

Total 4 crore cases are pending out of which 73% are criminal and 27% are civil cases

### Finding pending cases in thousands and storing in new column

```python
court_df['pending_thousands'] = court_df['Pending cases']/1000

court_df.head(10)
```

Created a truncated dataset by removing the total values for all states/districts, and created new columns for pending cases in thousands

```python
Court_case_type = court_df.groupby('Court_case_type')['pending_thousands'].agg('sum').reset_index()

Court_case_type
```

```python
dat1 = plt.pie(Court_case_type['pending_thousands'],labels=Court_case_type['Court_case_type'],autopct='%1.1f%%')
plt.title('Break-up of case type in thousands', bbox={'facecolor':'0.8', 'pad':5})
dat1
```

### Percentage of cases pending in courts for over 10 years, and the variation (if any) at a state level (in the code below)

```python
court_df['pending_over_10_yrs'] = court_df['Pending cases for a period of 10 to 20 years']+court_df['Pending cases for a period of 20 to 30 years']+court_df['Pending cases over 30 years']
court_df['pending_over_10_yrs_thousands'] = court_df['pending_over_10_yrs']/1000

court_df.head(10)
```

### Number of pending cases in thousands statewise

```python
states_pending_total = court_df.groupby('State')['pending_thousands'].agg('sum').reset_index()

states_pending_total.head(10)
```

### Number of pending cases over 10 years in thousands

```python
states_10yrs_total = court_df.groupby('State')['pending_over_10_yrs_thousands'].agg('sum').reset_index()

states_10yrs_total.head(10)
```

```python
states_10yrs_combb = pd.merge(states_pending_total,states_10yrs_total,on='State')

states_10yrs_combb.head(10)
```

```python
states_10yrs_combb['perc_over_10yrs'] =states_10yrs_combb['pending_over_10_yrs_thousands']/states_10yrs_combb['pending_thousands']*100

states_10yrs_combb.head(10)
```

```python
states_10yrs_combb1 =states_10yrs_combb.groupby('State')['perc_over_10yrs'].agg('mean').sort_values(ascending=False).reset_index()

states_10yrs_combb1.head(10)
```

```python
dat2 = plt.bar(states_10yrs_combb['State'],states_10yrs_combb['perc_over_10yrs'])
plt.xticks(color='black',rotation=90,fontsize='8', horizontalalignment='right')
plt.tick_params(labelbottom=True)
```

##### Poorer and resource constrained states (UP, WB, Bihar, Jharkhand, Rajasthan) are at the top of the distribution whereas the better off states (mostly southern, and north eastern) are at the tail end which means that economic condition and resource availability in a state also have an impact on the case pendency in that state

```python
state_Court_case_type = court_df.groupby(['State','Court_case_type'])['pending_thousands'].agg('sum').reset_index()

state_Court_case_type.head(10)
```

```python
state_wise_case_pivot = pd.pivot_table(data=state_Court_case_type,values='pending_thousands',index='State',columns='Court_case_type').reset_index()

state_wise_case_pivot.head(10)
```

```python
state_wise_case_pivot['criminal_pcnt'] = (state_wise_case_pivot['Criminal']/(state_wise_case_pivot['Civil']+state_wise_case_pivot['Criminal']))*100

state_wise_case_pivot['civil_pcnt'] = 100-state_wise_case_pivot['criminal_pcnt']

state_wise_case_pivot.head(10)
```

```python
state_wise_case_pivot1 = state_wise_case_pivot.groupby('State')['criminal_pcnt'].agg('mean').sort_values(ascending=False).reset_index()

state_wise_case_pivot1.head(10)
```

```python
X =state_wise_case_pivot['State']
X_axis = np.arange(len(X))
plt.bar(X_axis - 0.2,state_wise_case_pivot['criminal_pcnt'],0.5, label = 'Criminal Cases')
plt.bar(X_axis + 0.2,state_wise_case_pivot['civil_pcnt'],0.5, label = 'Civil Cases')
plt.tick_params(labelbottom=True)
plt.rcParams["figure.figsize"] = [8.50, 8.50]
plt.xticks(X_axis, X,color='black',rotation=90,fontsize='8', horizontalalignment='right')
plt.title("Pending cases in thousand state wise on the basis of nature of cases")
plt.legend()
plt.show()
```

Distribution of civil and criminal cases is same in all states. Criminal cases are more in the pending cases again, the economically weak states again show the high pendencies of cases.

```python
state_comb2 = pd.merge(states_10yrs_combb1,state_wise_case_pivot1,on='State')
state_comb2
```

```python
sns.scatterplot(x=state_comb2.perc_over_10yrs,y=state_comb2.criminal_pcnt)
```

Although not perfect, there does seem to be some correlation between high % of criminal cases and high delay in dispensing a verdict. It would be interesting to see the reasons for high pendencies to better understand this dynamic

```python
pend_evidence = court_df.groupby('Court_case_type')['Cases pending at evidence or argument or judgement stage'].agg('sum').reset_index()

pend_evidence
```

```python
pending_compliance = court_df.groupby('Court_case_type')['Cases pending at compliance or steps or stay stage'].agg('sum').reset_index()

pending_compliance
```

```python
pending_appearance = court_df.groupby('Court_case_type')['Cases pending at appearance or service-related stage'].agg('sum').reset_index()

pending_appearance
```

```python
pending_charge = court_df.groupby('Court_case_type')['Cases pending at pleadings or issues or charge stage'].agg('sum').reset_index()

pending_charge
```

```python
pend_1 = pd.merge(pend_evidence,pending_compliance,on='Court_case_type')

pend_1
```

```python
pend_2 = pd.merge(pending_appearance,pending_charge,on='Court_case_type')

pend_2
```

```python
pend_3 = pd.merge(pend_1,pend_2,on='Court_case_type')

pend_3
```

```python
X =pend_3['Court_case_type']
X_axis = np.arange(len(X))
plt.bar(X_axis - 0.3,pend_3['Cases pending at evidence or argument or judgement stage'],0.2, label = 'ases pending at evidence or argument or judgement stage')
plt.bar(X_axis - 0.1,pend_3['Cases pending at compliance or steps or stay stage'],0.2, label = 'Cases pending at compliance or steps or stay stage')
plt.bar(X_axis + 0.1,pend_3['Cases pending at appearance or service-related stage'],0.2, label = 'Cases pending at appearance or service-related stage')
plt.bar(X_axis + 0.3,pend_3['Cases pending at pleadings or issues or charge stage'],0.2, label = 'Cases pending at pleadings or issues or charge stage')
plt.tick_params(labelbottom=True)
plt.rcParams["figure.figsize"] = [8.50, 8.50]
plt.xticks(X_axis, X,color='black',rotation=90,fontsize='8', horizontalalignment='right')
plt.title("Pending cases in thousand state wise on the basis of nature of cases")
plt.legend()
plt.show()
```

A somewhat clear picture begins to appear here

(i) The most frequent reason for criminal cases pendency is at an appearance or service-related stage, which require the capability of the state to do so. Hence, states lacking in resources have a worst track record in case pendency of criminal.
(ii) For civil cases, the msot frequent reason for delay is at an evidence or argument stage which can be termed a 'technical delay issue' and has less to do with capacity.

States with extremely high pendencies seem to be those which have a high percentage of criminal cases where delay is due to capacity issued which is states with less financial resources.
